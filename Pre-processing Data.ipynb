```python
#Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading data from URL 
filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
filename
```

```python
#Adding the related headers
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename)
df.head()
```

```python
df.columns = headers
df.head(10)
```

```python
#Replace "?" to NaN

df.replace("?",np.nan, inplace = True)
df.head(5)
```

```python
#Evaluating for missing data

missing_data = df.isnull()
missing_data.head()
```

```python
#Count missing values in each column

for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
#Calculate the mean value for the "normalized-losses" column
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print('Average of normalized-losses :',avg_norm_loss)
```

```python
#Replace "NaN" with mean value in "normalized-losses" column
df["normalized-losses"].replace(np.nan,avg_norm_loss, inplace = True)
df.head()
```

```python
#Calculate the mean value for the "bore" column

avg_bor = df["bore"].astype('float').mean(axis =0)
print("Avrage of bore :",avg_bor)
```

```python
#Replace "NaN" with the mean value in the "bore" column
df["bore"].replace(np.nan, avg_bor, inplace = True)
df.head()
```

```python
# Calculate the mean value for the "stroke" column

avg_str = df["stroke"].astype("float").mean(axis = 0)
print("Avrege of stroke column :",avg_str)

#Replace "NaN" with the mean value in the "stroke" column
df["stroke"].replace(np.nan, avg_str, inplace = True)
```

```python
#Calculate the mean value for the "horsepower" column

avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)

#Replace "NaN" with the mean value in the "horsepower" column
df["horsepower"].replace(np.nan, avg_str, inplace = True)
```

```python
#Calculate the mean value for the "peak-rpm" column

avg_horsepower = df['peak-rpm'].astype('float').mean(axis=0)
print("peak-rpm:", avg_horsepower)

#Replace "NaN" with the mean value in the "horsepower" column
df["peak-rpm"].replace(np.nan, avg_str, inplace = True)
```

```python
#See which values are present in a particular column

df['num-of-doors'].value_counts()
```

```python
#Calculate the most common type automatically
df['num-of-doors'].value_counts().idxmax()
```

```python
#Replace the missing 'num-of-doors' values by the most frequent
df['num-of-doors'].replace(np.nan, "four", inplace = True)
```

```python
#Simply drop whole row with NaN in "price" column
df.dropna(subset = ['price'], axis = 0, inplace = True)

#Reset index
df.reset_index(drop = True, inplace = True)
```

```python
df.head(10)
```

```python
#The data types
df.dtypes
```

```python
#Convert data types to proper format
df[['bore','stroke','price','peak-rpm']] = df[['bore','stroke','price','peak-rpm']].astype('float')
df[['normalized-losses']] = df[['normalized-losses']].astype('int')
```

```python
df.dtypes
```

```python
#Data transformation
#Convert mpg to L/100km by mathematical operation(235 divided by mpg)
df['city-L/100km'] = 235/df['city-mpg']
df.head()
```

```python
df['highway-mpg'] = 235/df['highway-mpg']
#Change the name of the column
df.rename(columns = {'highway-mpg':'highway-L/100km'}, inplace = True)
df.head()
```

Data normalization

```python
#Replace (original value) by(original value)/(maximum value)
df['length'] = df['length']/df['length'].max()
df[['width','height']] = df[['width','height']]/df[['width','height']].max()
```

```python
df[['length','width','height']].head()
```

Binning

```python
#Convrt data to correct format
df["horsepower"] = df["horsepower"].astype(int, copy=True)
df["horsepower"].head()
```

```python
#Plot histogram
plt.hist(df["horsepower"])

#set x/y labels and plot title
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
```

```python
bins = np.linspace(min(df["horsepower"]), max(df['horsepower']),4)
bins
```

```python
group_names = ['Low','Medium','High']
df['horsepower-binned'] = pd.cut(df['horsepower'],bins, labels = group_names, include_lowest = True)
df[['horsepower','horsepower-binned']].head(20)
```

```python
df["horsepower-binned"].value_counts()
```

```python
plt.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
```

```python
# draw historgram of attribute "horsepower" with bins = 3
plt.hist(df["horsepower"], bins = 3)

# set x/y labels and plot title
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
```
